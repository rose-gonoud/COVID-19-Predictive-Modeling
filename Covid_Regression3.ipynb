```python
# Read in CSV files
covid = pd.read_csv('../COVID-19-Predictive-Modelling/assets/data/COVID County Data/covid_county_data_cleaned2.csv')
social = pd.read_csv('../COVID-19-Predictive-Modelling/assets/data/SVI2018_US_COUNTY_2.csv')
employment = pd.read_csv('../COVID-19-Predictive-Modelling/assets/data/AGS_data/AGS_Data_Massaged2.csv')
```

```python
covid.fips = covid.fips.astype('str')
covid.head()
```

```python
social.fips = social.fips.astype('str')
social.head()
```

```python
# For the SVI, -999 reflects missing values. Apparantly, only one county has -999 vaules in their dataset, so removing.
social =  social[(social.iloc[:, 1:] != -999).all(axis=1)]
social.head()
```

```python
fips_grouped = covid.groupby('fips')[['confirmed', 'deaths', 'confirmed_diff', 'deaths_diff']].sum()
print(fips_grouped)
```

```python
flat_fips = fips_grouped.reset_index()
print(flat_fips)
```

```python
flat_fips.head()
```

```python
# Merge flat_fips and social on fips
merged = pd.merge(social, flat_fips, on='fips')
merged.head()
```

```python
merged.dtypes
```

```python
merged.fips = merged.fips.astype('int')
merged.head()
```

```python
employment.dtypes
```

```python
# Merge merged and employment on fips
combined = pd.merge(merged, employment, on='fips')
combined.head()
```

```python
combined.shape
```

```python
# Note that total population in these calculations is from the SVI, which was conducted 2014-2018. So, it's not the current number, but an approximation. 
combined['percent_confirmed'] = ((combined['confirmed'] / combined['E_TOTPOP'].sum())*100).round(10).astype(str)
```

```python
combined['percent_deaths'] = ((combined['deaths'] / combined['E_TOTPOP'].sum())*100).round(10).astype(str)
```

```python
combined.head()
```

```python
# Model 1: Linear regression using confirmed (summed count) from SVI
X = combined[['AREA_SQMI','E_HU', 'E_HH', 'E_POV', 'E_UNEMP', 'E_PCI', 'E_NOHSDP', 'E_AGE65', 'E_AGE17', 'E_DISABL', 'E_SNGPNT', 'E_MINRTY', 'E_LIMENG', 'E_MUNIT', 'E_MOBILE', 'E_CROWD', 'E_NOVEH', 'E_GROUPQ']]
```

```python
Y = combined['confirmed'].values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X, Y)
```

```python
score = model.score(X, Y)
print(f"R2 Score: {score}")
```
